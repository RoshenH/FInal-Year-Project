```python
import os
from astropy.io import fits
import pandas as pd

# Directory containing your FITS files
fits_dir = r"C:\Users\Roshen Hasangha\Desktop\dataset\new folder"

# Ensure the directory exists (optional, since it should)
if not os.path.exists(fits_dir):
    print(f"Error: Directory {fits_dir} does not exist.")
    exit(1)

# Loop through all FITS files in the directory
for fits_file in os.listdir(fits_dir):
    if fits_file.endswith(".fits"):
        fits_path = os.path.join(fits_dir, fits_file)
        try:
            # Open the FITS file
            with fits.open(fits_path) as hdul:
                data = hdul[1].data
                # Extract time, flux, and flux error
                df = pd.DataFrame({
                    "Time": data["TIME"],  # Barycentric Julian Date (BJD)
                    "Flux": data["PDCSAP_FLUX"],  # Corrected flux
                    "Flux_Error": data["PDCSAP_FLUX_ERR"]
                })
                # Save to CSV with the same base name
                csv_file = fits_file.replace(".fits", ".csv")
                df.to_csv(os.path.join(fits_dir, csv_file), index=False)
                print(f"Converted {fits_file} to {csv_file}")
        except Exception as e:
            print(f"Error converting {fits_file}: {e}")

print("Conversion process completed.")
```
